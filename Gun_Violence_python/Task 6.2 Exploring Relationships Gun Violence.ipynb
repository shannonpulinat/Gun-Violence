## 6. Categorical Plots

```python
# Histogram for Top 10 Cities with maximum incidents of Gun Violence
citywise_total = df_gunv[["incident_id"]].groupby(df_gunv["city"]).count()
top_cities = citywise_total.sort_values(by='incident_id', ascending=False).head(10)
print(top_cities)
top_cities.plot.barh()
del(top_cities)
```

```python
# Histogram for Top 10 States with maximum incidents of Gun Violence
statewise_total = df_gunv[["incident_id"]].groupby(df_gunv["state"]).count()
top_states = statewise_total.sort_values(by='incident_id', ascending=False).head(10)
print(top_states)
top_states.plot.barh()
del(top_states)
```

```python
# Histogram for Weekday wise Incidents
weekwise_total = df_gunv[["incident_id"]].groupby(df_gunv["day_of_week"]).count()
weekwise_total.plot.barh()
del(weekwise_total)
# Weekdays 0 is Monday and 6 is Sunday.
```

```python
# Histogram for Yearly Incidents
year_total = df_gunv[["incident_id"]].groupby(df_gunv["year"]).count()
year_total.plot.barh()
del(year_total)
```

```python
# Histogram for Month wise Incidents
monthwise_total = df_gunv[["incident_id"]].groupby(df_gunv["month"]).count()
monthwise_total.plot.barh()
del(monthwise_total)
```
